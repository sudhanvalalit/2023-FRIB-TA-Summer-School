# Empirical Interpolation Challenge

(adapted from Kyle Beyer's tutorial http://rbm.ascsn.net/scattering/eim.html)

The purpose of this challenge is for you to build a reduced basis approximation for an operator, in this case a Woods-Saxon potential:

\begin{equation}
    U(r;V_0,R,a) = -V_0\frac{1}{1 + \exp{\frac{r-R}{a}}}.
\end{equation}

This potential (or operator) depends on three parameters $\alpha=(V_0,R,a)$, with a non-affine dependence in $R$ and $a$ which could make some RBM applications less efficient. Our objective here is to expand it as:

\begin{equation}
U(r; \alpha)  \approx \tilde{U}_\text{EIM} (r; \alpha)  = \sum_k^M \beta_k(\alpha) U_k(r),
\end{equation}

where the $M$ functions $U_k(r)$ are fixed and only the coefficients $\beta_k$ depend on $\alpha$. We could build the basis $U_k(r)$ from a Principal Component Analysis (or Singular Value Decomposition), while the coefficients $\beta_k$ are obtained for each new value of $\alpha$ through interpolation:

\begin{equation}
U(r_j; \alpha) - \sum_k^M \beta_k(\alpha) U_k(r_j)=0, \ \text{for } j=1, ... , M
\end{equation}


Your tasks are:

- Build a reduced basis for the potential by principal component analyis of a handful of values. Good central values could be V0=40, R=4, a=0.5. You can also compare to http://rbm.ascsn.net/scattering/eim.html.
- Build a function that given a new value of the parameters creates the interpolating equations and solves for the RB coefficients.
- Compare the accuracy of your reduced basis implementation using some metric (perhaphs the L2 norm).
- Bonus: integrate this method with a reduced basis approach to the wave functions as well.

```python
import numpy as np

import matplotlib.pyplot as plt

import time

import timeit

```

```python
x = np.linspace(0.001,15, 300)
```

```python
def woods_saxon(r, args):
    V0, R, a = args
    return -V0/(1. + np.exp((r-R)/a))
    
```

```python
fig = plt.figure(dpi=150)

plt.plot(x,woods_saxon(x, [40,5,0.5]))

plt.show()
```
